```python
# https://www.sfu.ca/~ssurjano/schwef.html
def Schwefel(x,y):
    n =2;
    V = 4189.829101;

#     x = np.linspace(-500,100,500)

#     y = np.linspace(-500,100,500)

#     [x,y] = np.meshgrid(x,y)

    z = n*V+(-x*np.sin(np.sqrt(np.abs(x)))) + (-y*np.sin(np.sqrt(np.abs(y))))
    # print(z)
    return z
```
